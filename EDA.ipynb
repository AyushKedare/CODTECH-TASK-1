```python
fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(14,10))

# 1. Bar Plot: Top 10 Products by Orders
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar', ax=ax1)
ax1.set_title('Top 10 Products by Orders')
ax1.set_xlabel('Product ID')
ax1.set_ylabel('Total Orders')

# 2. Scatter Plot: Add a scatter plot (example: Orders vs Product_ID)
ax2.scatter(df['Amount'], df['Orders'], alpha=0.5)
ax2.set_title('Scatter Plot: Orders vs Product ID')
ax2.set_xlabel('Amount')
ax2.set_ylabel('Orders')

# 3. Histogram: Distribution of Orders
ax3.hist(df['Orders'], bins=10, color='green', alpha=0.7)
ax3.set_title('Histogram: Distribution of Orders')
ax3.set_xlabel('Orders')
ax3.set_ylabel('Frequency')

# 4. Heatmap: Correlation matrix heatmap
# Select only numeric columns for the correlation matrix
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
corr_matrix = numeric_df.corr()

# Plot the heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', ax=ax4)
ax4.set_title('Heatmap: Correlation Matrix')

# Adjust layout for better readability
plt.tight_layout()

# Display the plots
plt.show()
```
